```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y)
plt.title('First Graph', fontdict={'fontname' : 'Comic Sans MS'})
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])



plt.show()
```
